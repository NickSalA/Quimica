## Pregunta 1: DataFrames y listas en python (4 puntos)

Usando PeriodicTable, crea una tabla con 20 elementos químicos y en las columnas considera: nombre, simbolo, masa_atomica, numero_atomico	y valencias. Puedes ayudarte del siguiente [material](https://github.com/inefable12/UNMSM_quimica_2023/blob/main/Python_para_quimica_2023.ipynb).

```python
!pip install rdkit
```

```python
from rdkit import Chem
PeriodicTable = Chem.GetPeriodicTable()
elements = []

for element in range(1,21):
  elements.append(PeriodicTable.GetElementSymbol(element))
print(elements)

NumeroAtomico = [PeriodicTable.GetAtomicNumber(x) for x in elements]
MasaAtomica = [PeriodicTable.GetAtomicWeight(x) for x in elements]
NombreDelElemento = [PeriodicTable.GetElementName(x) for x in range(1,21)]
Valencia = [PeriodicTable.GetValenceList(x) for x in range(1,21)]

```

```python
import pandas as pd

data = {'Nombre': NombreDelElemento, 'Simbolo': elements,
        'MasaAtomica': MasaAtomica, 'NumeroAtomico': NumeroAtomico,
        'Valencia': Valencia}
df = pd.DataFrame(data)
df.index=df.index+1
df
```

```python
from sympy import Eq, solve, symbols

b, c, d = symbols('b c d')

C = Eq(4, c)
H = Eq(8, 2*d)
O = Eq(2*b, 2*c + d)

solution = solve((C, H, O), (b, c, d))

coefficients = [solution[b], solution[c], solution[d]]

ReacciónQuimica = f"1 C4H8 + {coefficients[0]} O2 -> {coefficients[1]} CO2 + {coefficients[2]} H2O"
print("Reaccion Química balanceada:")
print(ReacciónQuimica)

print('''
Los subíndices o moleculariedad son los siguientes:
En C4H8: 1 C4H8 (cuatro átomos de carbono y ocho átomos de hidrógeno).
En O2  : 4 O2   (Ocho átomos de oxígeno).
En CO2 : 4 CO2  (Cuatro átomos de carbono y ocho átomos de oxígeno).
En H2O : 4 H2O  (Ocho átomos de hidrógeno y cuatro átomos de oxígeno).''')
```

#El metodo cientifico

El método científico desempeña un papel fundamental en el avance de la ciencia y la ampliación del conocimiento y comprensión de la realidad. Se logra a través de la formulación de teorías y leyes. Sin embargo, la comprensión de esta vasta cantidad de información puede resultar altamente compleja. Este desafío, especialmente evidente en el ámbito educativo, plantea la pregunta fundamental: ¿Cómo facilitar la comprensión de los diversos conceptos y descubrimientos científicos para estudiantes de secundaria o incluso universitarios?

Para abordar esta cuestión, los investigadores han identificado un problema específico y delimitaron su objetivo: ¿Cómo lograr que los estudiantes comprendan la configuración electrónica y su relación con las propiedades y periodicidad de manera efectiva? Una solución ingeniosa y particular surgió en este proceso. Se utilizaron cajas de huevos como contenedores para representar a los protones y neutrones, aspectos que a menudo no reciben suficiente atención en las explicaciones sobre configuración electrónica. Por otro lado, se puso un fuerte énfasis en los electrones, representados por bolas de ping pong y líneas que indicaban los spins, dispuestos siguiendo las reglas de Moeller y el principio de Hund. Este enfoque permitió a los alumnos comprender a fondo los diversos conceptos relacionados con la configuración electrónica, lo cual se reflejó en mejoras evidentes en las representaciones gráficas.

En resumen, el uso del método científico condujo al desarrollo de una solución que resulta económicamente viable y altamente efectiva para abordar el desafío planteado.
